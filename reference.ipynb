```python
# Simple chatbot program

precepts = ["hello", "tea", "cafe", "no", "thank you", "goodbye"]
# Define a list of known questions and their responses

responses = [
    "Hello! How can I assist you today?",
    "Sure, where would you like to it? A cafe in the campus maybe?",
    "There are 2 cafes - in main building and near the lecture halls.",
    "Is there anything else I can help you with",
    "You are welcome.",
    "Goodbye! Have a great day!"
]
```

```python
# Function to get user input and respond
def chatbot():

    # User welcome and termination instruction.

    # Main loop:
    while True:
        
        # 1. Get user input
        user_input = input("You: ")
                
        # 2. Check if the user wants to exit
        if user_input == "bye":
            print("Chatbot: Thank you for your interaction")
            break
         
        # 3. Search for a pattern in the precepts. And fetch a predefined responses
        response_found = False
        for pattern in precepts:
            
            if pattern in user_input:
                response_found = True
                ind = precepts.index(pattern)
                response = responses[ind]
                print("Chatbot: ", response)
                break;
        
        if response_found == False:
            print("I'm sorry, could you rephrease that again")
                
        #   3.1 Check if any of precepts are there in the input
        #   3.2 Return the index
        #   3.3 Use index to find repsonse
        #   3.4 Break the loop
        
# Run the chatbot
chatbot()
```

### 2.1 Lists -> Dictionaries

To convert lists to dictionaries, let's use the lists from previous activity.

```python
import re
import random

# Write a function that takes in precept categories and responses lists and returns a mapping between them.

# - Use ZIP function to combine two independent lists (precepts and responses) and create a new map between them.
# - Search on the ZIP function and how to use it in a for loop
# - Run the main loop and check that this function works

# ___ Write your code here _____

def zip_responses(precepts, responses):
    mapping = {}
    for precept, response in zip(precepts, responses):
        mapping[precept] = response
    return mapping
```

### 2.3 Function : Detect pattern from input

---

Use regular expressions + dictionary from before to detect a pattern from user input. 

**Function input** : User Input ("Hi, how are you?")  <br>
**Output**         : Response_type ("Greetings")

```python
## Function name: "detect_pattern"

## Create the detect_pattern function with one argument : user_input
## 1. Loop through each key in the dictionary:
## 2. Write if-else conditions to check for pattern match
##    2.1 For the if condition, use the re.search to check if the pattern exists in user input 
##    2.2 If there is a match, return the response type (e.g., "greetings" or "bye" or "happy")
##    2.3 If none of the conditions match, return "default" which will be the default response of the chatbot
##    Note: use else for last conditon

## TIP: To avoid multiple conditions from matching, use a break statement. 

# Give the function a characteristic name (e.g., detect_pattern) 
# ___ Write your code here _____

def detect_pattern(user_input):
    for pattern, response_type in re_patterns.items():
        if re.search(pattern, user_input, re.IGNORECASE):
            return response_type
        
    return "default"
```

### 2.6 Advanced regular expressions

Let's exploring advanced REs by creating two more patterns to match the name and favourite color of the user. 

As name, and color can be any value, pattern matching needs to be generic and requires using alphabets [a-zA-Z], words [\w] and special symbols [*,+,?].

```python
import re

# Write a regular expression pattern for a name and favourite color.

# 1. Names typically start with a letter in upper case followed by a sequence of letters in lower case.
# Try writing a few use cases to ensure that the pattern works.

name_pattern = r"my name is ([A-Z][a-z]+)"

# 2. Colors typically start with a pattern - color is <color_name> or I like <color_name>.
# Try writing a few use cases to ensure that the pattern works.

color_pattern = r"(?:color is|I like) ([a-zA-Z]+)"
```

```python
# ___ Write your code here _____

# Write few examples of input strings containing names and color to ensure that the pattern works.

name_inputs = [
    "my name is Alice",
    "my name is Bob",
    "my name is charlie",   # won't match - lowercase
    "my name is DAVE",      # won't match - all uppercase
]

for test in name_inputs:
    result = re.search(name_pattern, test)
    if result:
        print(f"'{test}' => Match: {result.group(1)}")
    else:
        print(f"'{test}' => No match")

color_inputs = [
    "my color is blue",
    "I like green",
    "color is red",
    "I like YELLOW",
    "my favourite is purple",  # won't match - different phrasing
]

for test in color_inputs:
    result = re.search(color_pattern, test, re.IGNORECASE)
    if result:
        print(f"'{test}' => Match: {result.group(1)}")
    else:
        print(f"'{test}' => No match")
```

```python
import re
import random

re_patterns = {
    r"hi|hello|hey|howdy|greetings|yo|sup":         "greetings",
    r"tea|coffee|drink|beverage|latte|espresso":     "tea",
    r"cafe|coffee shop|canteen|restaurant|eatery":   "cafe",
    r"thanks|thank you|cheers|appreciated|grateful": "thanks",
    r"bye|goodbye|see you|farewell|later|ciao":      "goodbye"
}

chatbot_responses = {
    "greetings":      ["Hey there! What can I do for you?", "Hello! How can I help?"],
    "tea":            ["I love tea! Would you like a recommendation?", "Great choice! What kind do you prefer?"],
    "cafe":           ["We have two cafes — main building and near lecture halls.", "The cafe near the library is great!"],
    "thanks":         ["You're welcome!", "Happy to help!", "Anytime!"],
    "goodbye":        ["Goodbye! Have a great day!", "See you soon!"],
    "name_response":  ["Hello {name}, how are you today?", "Nice to meet you {name}!"],
    "color_response": ["Great to know that your favourite color is {color}!", "{color} is a wonderful choice!"],
    "default":        ["I'm not sure I understand. Could you rephrase that?", "Can you tell me more?"]
}

user_state = {
    "name":  None,
    "color": None
}
print("test...")
print(user_state)

name_pattern  = r"my name is ([A-Z][a-z]+)"
color_pattern = r"(?:color is|I like) ([a-zA-Z]+)"

def detect_pattern(user_input):
    # Check name pattern
    result = re.search(name_pattern, user_input)
    if result:
        user_state["name"] = result.group(1)
        return "name_response"

    # Check color pattern
    result = re.search(color_pattern, user_input, re.IGNORECASE)
    if result:
        user_state["color"] = result.group(1)
        return "color_response"

    # Check existing patterns
    for pattern in re_patterns.keys():
        if re.search(pattern, user_input, re.IGNORECASE):
            return re_patterns[pattern]

    return "default"

print("test...")
print(detect_pattern("my name is Alice"), "=> user_state:", user_state)
print(detect_pattern("I like blue"),      "=> user_state:", user_state)
print(detect_pattern("hello there"),      "=> user_state:", user_state)
print(detect_pattern("random input"),     "=> user_state:", user_state)

def chatbot_response(user_input):
    detected_type = detect_pattern(user_input)

    name  = user_state["name"]
    color = user_state["color"]

    if detected_type == "name_response":
        response = random.choice(chatbot_responses["name_response"])
        return response.format(name=name)

    elif detected_type == "color_response":
        response = random.choice(chatbot_responses["color_response"])
        return response.format(color=color)

    elif detected_type in chatbot_responses:
        return random.choice(chatbot_responses[detected_type])

    else:
        return random.choice(chatbot_responses["default"])

print("test...")
print(chatbot_response("my name is Alice"))
print(chatbot_response("I like blue"))
print(chatbot_response("hello"))
print(chatbot_response("random input"))

def chat():
    print("\nChatbot: Hello! How can I help you today?")
    while True:
        user_input = input("You: ")
        if re.search(r"bye|goodbye|exit|quit", user_input, re.IGNORECASE):
            print("Chatbot: Goodbye! Have a great day!")
            break
        bot_response = chatbot_response(user_input)
        print(f"Chatbot: {bot_response}")

chat()
```

### 3.3 Integrating file-based data into chatbot logic

Follow the code templates provided below to understand how files can be integrated with the Chatbot logic.

```python
## A code template for storing responses from JSON file..

# def load_files():
#    pd.read_json('intents.json')
#    Type your code here.

import pandas as pd
import re
import random

def load_files():
    intents = pd.read_json('intents.json')
    
    rgx2int = {}
    int2res = {}
    
    for intent in intents["intents"]:
        tag = intent["tag"]
        rgx2int[r"|".join(intent["patterns"])] = tag
        int2res[tag] = intent["responses"]
    
    return rgx2int, int2res
```

```python
## A template for what the response generation code will look like..

def generate_response(user_input):
    
    rgx2int, int2res = load_files()
    
    # Checks for user input in rgx2int
    # Uses tag of rgx2int in int2res to get output
    
    # Print match and return true
    # or return False 
    
    for pattern, tag in rgx2int.items():
        if re.search(pattern, user_input, re.IGNORECASE):
            responses = int2res[tag]
            print(random.choice(responses))
            return True
    
    return False

# Test
result = generate_response("Hello there")
if not result:
    print("I didn't understand that.")
```

```python
## A reminder for what the main loop code will look like..

# Function to get user input and respond
def chatbot():

    # User welcome and termination instruction.
    print("Welcome! Press bye to exit")

    # Main loop:
    while True:
        
        user_input = input("You: ")
                
        if user_input == "bye":
            print("Chatbot: Thank you for your interaction")
            break
        
        response_found = generate_response(user_input)
                
        if response_found == False:
            print("I'm sorry, could you rephrase that again")
        
# Run the chatbot
chatbot()
```

```python
# ___ Write your code here _____

# references:
# https://www.nltk.org/howto/stem.html
# https://www.nltk.org/api/nltk.stem.porter.html#nltk.stem.porter.PorterStemmer
# https://www.nltk.org/api/nltk.stem.wordnet.html#nltk.stem.wordnet.WordNetLemmatizer
# https://go.oreilly.com/university-of-london-worldwide/https:/learning.oreilly.com/library/view/text-analytics-with/9781484223871/

import nltk
import re
import random
import pandas as pd

nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

stop_words = stopwords.words('english')
ps = PorterStemmer()
wnl = WordNetLemmatizer()

# maps a penn treebank pos tag (from nltk.pos_tag) to the format WordNetLemmatizer understands.
def get_wordnet_pos(tag):
    if tag.startswith('V'):
        return 'v' # verb
    elif tag.startswith('J'):
        return 'a' # adjective
    elif tag.startswith('R'):
        return 'r' # adverb
    else:
        return 'n' # noun

def text_preprocessing(user_input, method="porter"):
    # tokenizes user input, removes punctuation and stopwords
    tokens = word_tokenize(user_input)
    no_punct = [word for word in tokens if word.isalpha()]
    no_stop = [word for word in no_punct if word.lower() not in stop_words]
    
    # applies word substitution via stemming or lemmatization
    # and returns a list of cleaned and normalized tokens
    if method == "porter":
        processed = [ps.stem(word) for word in no_stop]
    elif method == "lemmatize":
        tagged = pos_tag(no_stop) # part of speech tags for example ('better', 'RBR')
        processed = [wnl.lemmatize(word.lower(), pos=get_wordnet_pos(tag)) for word, tag in tagged]
    else:
        raise ValueError("method must be 'porter' or 'lemmatize'")
    return processed

def load_files(method="porter"):
    intents = pd.read_json('intents.json')
    
    rgx2int = {}
    int2res = {}
    
    for intent in intents["intents"]:
        tag = intent["tag"]

        # normalizes patterns in the same way user input is processed
        stemmed_patterns = []
        for p in intent["patterns"]:
            processed = text_preprocessing(p, method=method)
            
            if processed: # skip vanished patterns for example all stopwords
                stemmed_patterns.append(r"\b" + r"\s+".join(processed) + r"\b")
                
        if stemmed_patterns:
            rgx2int[r"|".join(stemmed_patterns)] = tag
            int2res[tag] = intent["responses"]
    
    return rgx2int, int2res

# stores past (user_input, response) pairs
memory = []

def generate_response(user_input, processed_input, method="porter"):
    # recall: what did I/you say
    if re.search(r"what did (i|you) (say|ask)", user_input, re.IGNORECASE):
        if memory:
            last_input, last_response = memory[-1]
            print(f"Chatbot: You said \"{last_input}\" and I replied \"{last_response}\"")
        else:
            print("Chatbot: We haven't talked about anything yet!")
        return True
    
    rgx2int, int2res = load_files(method=method)

    # matching processed input against intents
    joined_input = " ".join(processed_input)
    for pattern, tag in rgx2int.items():
        if re.search(pattern, joined_input, re.IGNORECASE):
            response = random.choice(int2res[tag])
            print("Chatbot:", response)
            memory.append((user_input, response)) # remember this input and response
            return True
    return False

def chatbot():
    print("Welcome! I'm a chatbot for the NLP stemming/lemmatization class topic.")
    print("Press bye to exit")
    
    method = "lemmatize" # porter or lemmatize
    
    while True:
        user_input = input("You: ")
                
        if user_input == "bye":
            print("Chatbot: Thank you for your interaction")
            break
        
        processed_tokens = text_preprocessing(user_input, method=method)
        response_found = generate_response(user_input, processed_tokens, method=method)
                
        if response_found == False:
            print("Chatbot: I'm sorry, could you rephrase that again")

chatbot()
```
