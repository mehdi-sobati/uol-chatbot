#### Sample conversation snippet:

> User: Hello

> Chatbot: Hello! Welcome to the Iran Quiz! What is your name?

> User: my name is Student

> Chatbot: Great! Ready to test your knowledge about Iran?

```python
# Precepts: keywords the chatbot listens for
precepts = ["hello", "hi", "name", "capital", "language", "currency", "saffron", "thanks", "bye", "exit", "quit"]

# Responses: one response per precept, matched by list index
responses = [
    "Hello! Welcome to the Iran Quiz! What is your name?",
    "Hi there! Welcome to the Iran Quiz! What is your name?",
    "Great! Ready to test your knowledge about Iran?",
    "The capital of Iran is Tehran, a city of over 15 million people.",
    "The official language of Iran is Persian, also known as Farsi.",
    "The official currency of Iran is the Iranian Rial.",
    "Iran produces around 90% of the world's saffron supply!",
    "You're welcome! Keep exploring Iran trivia.",
    "Goodbye! Thanks for playing the Iran Quiz!"
]
```

```python
def chatbot():
    print("Welcome to the Iran Quiz Chatbot!")
    print("Type 'exit' or 'quit' to end the chat.\n")

    while True:
        user_input = input("You: ").lower()

        # Check if the user wants to exit
        if user_input in ["exit", "quit", "bye"]:
            print("Chatbot: Goodbye! Thanks for visiting the Iran Quiz!")
            break

        # Search for a keyword in precepts and return the matching response
        response_found = False
        for precept in precepts:
            if precept in user_input:
                ind = precepts.index(precept)
                print("Chatbot:", responses[ind])
                response_found = True
                break

        if not response_found:
            print("Chatbot: I'm not sure about that. Try asking about the capital, language, or currency of Iran!")

chatbot()
```

```python
import re
import random

# Write a function that takes in precept categories and responses lists and returns a mapping between them.

# - Use ZIP function to combine two independent lists (precepts and responses) and create a new map between them.
# - Search on the ZIP function and how to use it in a for loop
# - Run the main loop and check that this function works

# ___ Write your code here _____

def zip_responses(precepts, responses):
    mapping = {}
    for precept, response in zip(precepts, responses):
        mapping[precept] = response
    return mapping
```

### 2.3 Function : Detect pattern from input

---

Use regular expressions + dictionary from before to detect a pattern from user input. 

**Function input** : User Input ("Hi, how are you?")  <br>
**Output**         : Response_type ("Greetings") 

---

```python
# Hardcoded quiz questions for Part 2
# Part 3 will load these from intents.json
quiz_questions = [
    {
        "question": "Q1: What is the capital city of Iran?",
        "answer": "tehran",
        "aliases": ["teheran"],
        "hint": "It starts with 'T', at the foot of the Alborz mountains."
    },
    {
        "question": "Q2: What is the official language of Iran?",
        "answer": "persian",
        "aliases": ["farsi"],
        "hint": "Also known by a name starting with 'F'."
    },
    {
        "question": "Q3: What ancient ceremonial capital of Iran was built around 515 BC?",
        "answer": "persepolis",
        "aliases": [],
        "hint": "Its name means 'City of Persians' in Greek."
    },
]

user_state = {
    "name": None,
    "score": 0,
    "current_question": 0,
    "answered": 0,
    "last_guess": None,
}

name_pattern = r"(?:my name is|i am|call me|i'm)\s+(?P<name>[A-Z][a-z]+)"
guess_pattern = r"(?:i think (?:it'?s?|the answer is)|maybe|is it|could it be|i'?d say|my answer is|i guess|it'?s?)?\s*(?P<guess>[A-Za-z][A-Za-z\s\-]{1,30})"

def detect_pattern(user_input):
    # 1. Named group: capture the user's name
    result = re.search(name_pattern, user_input)
    if result:
        user_state["name"] = result.group("name")
        return "name_response"

    # 2. Side-intent check first — so "hint", "score" etc. are never mistaken for a guess
    for pattern, intent in re_patterns.items():
        if re.search(pattern, user_input, re.IGNORECASE):
            return intent

    # 3. Named group: quiz guess — only when quiz is active (last resort)
    if user_state["name"] and user_state["current_question"] < len(quiz_questions):
        # strip trailing qualifiers so "Tehran i guess" → "Tehran"
        cleaned = re.sub(r"\s*\b(i guess|right\??|i think so|you know)\b\s*$", "", user_input, flags=re.IGNORECASE)
        # strip apostrophes — they break [A-Za-z\s\-] and split "it's" into "it" + "s"
        cleaned = re.sub(r"'", "", cleaned)
        result = re.search(guess_pattern, cleaned, re.IGNORECASE)
        if result:
            user_state["last_guess"] = result.group("guess").strip()
            return "quiz_guess"

    return "default"

# Test — "afaik it's Tehran" should now extract a guess containing "Tehran"
print(detect_pattern("my name is Student"), "| name:", user_state["name"])
print(detect_pattern("I think it's Tehran"), "| guess:", user_state["last_guess"])
print(detect_pattern("it's Tehran i guess"), "| guess:", user_state["last_guess"])
print(detect_pattern("afaik it's Tehran"), "| guess:", user_state["last_guess"])
print(detect_pattern("hint please"))
print(detect_pattern("what is my score"))
```

### 2.4 Function : Retrieve response

---

Use the output of pattern detection function (above) to trigger a chatbot response. 

In the function, use the "responses" dictionary from above. 


Follow the steps to write function. The function returns a response (e.g., "hello", or "how are you").

**Function input** : User Input ("Hi, how are you?")  <br>
**Detect pattern** : Response_type ("Greetings")  <br>
**Output**         : One of ["hello", "how are you?", etc...]

---

```python
def check_answer(guess, question):
    # return True if guess matches the correct answer or any accepted alias
    correct = [question["answer"]] + question["aliases"]
    return any(c in guess.lower() for c in correct)

def chatbot_response(user_input):
    intent = detect_pattern(user_input)
    name = user_state["name"] or "friend"
    score = user_state["score"]
    q_idx = user_state["current_question"]
    total = len(quiz_questions)

    if intent == "name_response":
        response = random.choice(chatbot_responses["name_response"]).format(name=name, total=total)
        return response + "\n" + quiz_questions[0]["question"]

    elif intent == "quiz_guess":
        question = quiz_questions[q_idx]
        if check_answer(user_state["last_guess"], question):
            user_state["score"] += 1
            feedback = random.choice(chatbot_responses["correct"]).format(name=name, score=user_state["score"])
        else:
            feedback = random.choice(chatbot_responses["wrong"]).format(name=name, answer=question["answer"])
        user_state["answered"] += 1
        user_state["current_question"] += 1
        next_idx = user_state["current_question"]
        if next_idx < total:
            return feedback + "\n" + quiz_questions[next_idx]["question"]
        else:
            return feedback + "\n" + random.choice(chatbot_responses["quiz_complete"]).format(
                name=name, score=user_state["score"], total=total)

    elif intent == "hint_response":
        if user_state["name"] and q_idx < total:
            return random.choice(chatbot_responses["hint_response"]).format(hint=quiz_questions[q_idx]["hint"])
        return random.choice(chatbot_responses["default"])

    elif intent == "score":
        return random.choice(chatbot_responses["score"]).format(
            name=name, score=score, answered=user_state["answered"])

    elif intent in chatbot_responses:
        return random.choice(chatbot_responses[intent])

    else:
        return random.choice(chatbot_responses["default"])

# Test
user_state.update({"name": None, "score": 0, "current_question": 0, "answered": 0, "last_guess": None})
print("---")
print(chatbot_response("my name is Student"))
```

```python
def chat():
    # Reset state for a fresh game
    user_state.update({"name": None, "score": 0, "current_question": 0, "answered": 0, "last_guess": None})

    print("Welcome to the Iran Quiz Chatbot!")
    print("Type 'exit' or 'quit' to stop at any time.\n")
    print("Chatbot: Hi there! What is your name?")

    while True:
        user_input = input("You: ")

        if re.search(r"\b(exit|quit|bye|goodbye)\b", user_input, re.IGNORECASE):
            name  = user_state["name"] or "friend"
            score = user_state["score"]
            total = len(quiz_questions)
            print(f"Chatbot: Goodbye {name}! Final score: {score}/{total}. See you next time!")
            break

        bot_response = chatbot_response(user_input)
        print(f"Chatbot: {bot_response}")

chat()
```

### 2.6 Advanced regular expressions

Let's create two named-group patterns: one to capture the user's name, and one to capture their quiz answer guess.

As names and answers can be any value, the patterns need to be flexible — using character classes `[a-zA-Z]`, quantifiers `+` and `?`, and non-capturing groups `(?:...)` to make prefixes optional.

The key feature is **named groups** `(?P<label>...)` — they let us extract matched text by a readable label: `result.group("name")` or `result.group("guess")` instead of `result.group(1)`.

```python
import re

# Named group for the user's name
name_pattern = r"(?:my name is|i am|call me|i'm)\s+(?P<name>[A-Z][a-z]+)"

# Named group for quiz answer guesses
# Optional prefix handles phrasings like "I think it's X", "maybe X", "it's X"
# Trailing qualifiers ("i guess", "right?") are stripped before this pattern is applied
guess_pattern = r"(?:i think (?:it'?s?|the answer is)|maybe|is it|could it be|i'?d say|my answer is|i guess|it'?s?)?\s*(?P<guess>[A-Za-z][A-Za-z\s\-]{1,30})"
```

```python
# Test name_pattern
name_tests = [
    "my name is Student",
    "My name is Mary",
    "I'm Gaurav",
    "call me Peter",
    "my name is max",    # no match — first letter must be uppercase
    "my name is DAVE",   # no match — all caps not matched
]
for test in name_tests:
    result = re.search(name_pattern, test)
    matched = result.group("name") if result else "No match"
    print(f"'{test}' => {matched}")

print()

# Test guess_pattern with various answer phrasings
guess_tests = [
    "Tehran",
    "I think it's Tehran",
    "maybe Persepolis",
    "is it Persian?",
    "could it be Farsi",
    "I'd say Damavand",
    "my answer is Rial",
    "7",
]
for test in guess_tests:
    result = re.search(guess_pattern, test, re.IGNORECASE)
    matched = result.group("guess").strip() if result else "No match"
    print(f"'{test}' => {matched}")
```

#### Integrating name and quiz guess into the chatbot

Now let's wire both named-group patterns into the full chatbot. Here's how the pieces connect:

1. `detect_pattern` checks `name_pattern` first — if matched, stores the name in `user_state["name"]` and returns `"name_response"`
2. `detect_pattern` then checks `guess_pattern`, but **only when the quiz is active** — stores the extracted guess in `user_state["last_guess"]` and returns `"quiz_guess"`
3. `chatbot_response` handles each intent — for `"quiz_guess"` it calls `check_answer()` to compare the guess against the correct answer and its aliases, then formats a template response using `{name}`, `{score}`, or `{answer}`

##### 1. Expanding `detect_pattern`
- Add `quiz_questions` list and `user_state` dict above the function
- Check `name_pattern` first using a named group — store result in `user_state["name"]`
- Check `guess_pattern` next, only when `user_state["name"]` is set and quiz is still running — store in `user_state["last_guess"]`
- Fall through to standard `re_patterns` matching, then return `"default"`

##### 2. Adding `check_answer` helper
- Compare `user_state["last_guess"]` against `question["answer"]` and `question["aliases"]`
- Return `True` if any match (case-insensitive)

##### 3. Expanding `chatbot_response`
- `"name_response"`: format greeting with `{name}` and `{total}`, append first question
- `"quiz_guess"`: call `check_answer()`, format right/wrong response, advance `current_question`, append next question or quiz-complete message
- `"hint_response"`: look up `quiz_questions[current_question]["hint"]` and format with `{hint}`
- `"score"`: format with `{name}`, `{score}`, `{answered}`

```python
import re
import random

# --- Data ---

intent_categories = [
    "greeting", "name_response", "correct", "wrong",
    "hint_response", "score", "quiz_complete", "goodbye", "default"
]

responses_list = [
    ["Hello! Welcome to the Iran Quiz! What is your name?",
     "Hey there! Ready for some Iran trivia? What's your name?"],
    ["Nice to meet you {name}! Let's begin with {total} questions about Iran.",
     "Great to have you {name}! Get ready for {total} Iran trivia questions."],
    ["Correct {name}! Well done. Your score is now {score}.",
     "That's right {name}! +1 point. Score: {score}."],
    ["Not quite {name}. The correct answer was: {answer}.",
     "Close but not right {name}. The answer was: {answer}."],
    ["Here's a hint: {hint}"],
    ["Your score so far: {score} out of {answered} answered, {name}.",
     "You've got {score}/{answered} right so far, {name}!"],
    ["Quiz complete! You scored {score} out of {total}, {name}. Well done!",
     "That's the end! Final score: {score}/{total}. Great effort {name}!"],
    ["Goodbye {name}! Final score: {score}/{total}. Come back soon!",
     "See you next time {name}! You scored {score}/{total}."],
    ["I'm not sure about that. Try again!", "Could you rephrase that?"]
]

def zip_responses(categories, responses):
    mapping = {}
    for category, response in zip(categories, responses):
        mapping[category] = response
    return mapping

chatbot_responses = zip_responses(intent_categories, responses_list)

# \b word boundaries prevent "hi" from matching inside "hint", "history", etc.
re_patterns = {
    r"\bhi\b|\bhello\b|\bhey\b|\bhowdy\b|\bgood morning\b|\bgreetings\b|\bsup\b": "greeting",
    r"\bscore\b|\bpoints\b|\bhow am i doing\b|\bhow many right\b|\bmy score\b": "score",
    r"\bhint\b|\bclue\b|\bhelp me\b|\bi don't know\b|\bno idea\b|\bskip\b|\bpass\b": "hint_response",
}

quiz_questions = [
    {"question": "Q1: What is the capital city of Iran?", "answer": "tehran", "aliases": ["teheran"], "hint": "It starts with 'T', at the foot of the Alborz mountains."},
    {"question": "Q2: What is the official language of Iran?", "answer": "persian", "aliases": ["farsi"], "hint": "Also known by a name starting with 'F'."},
    {"question": "Q3: What ancient ceremonial capital of Iran was built ~515 BC?", "answer": "persepolis", "aliases": [], "hint": "Its name means 'City of Persians' in Greek."},
]

user_state = {"name": None, "score": 0, "current_question": 0, "answered": 0, "last_guess": None}

name_pattern = r"(?:my name is|i am|call me|i'm)\s+(?P<name>[A-Z][a-z]+)"
guess_pattern = r"(?:i think (?:it'?s?|the answer is)|maybe|is it|could it be|i'?d say|my answer is|i guess|it'?s?)?\s*(?P<guess>[A-Za-z][A-Za-z\s\-]{1,30})"

# --- Functions ---

def detect_pattern(user_input):
    result = re.search(name_pattern, user_input)
    if result:
        user_state["name"] = result.group("name")
        return "name_response"

    # side-intents checked before quiz guess to prevent "hint" being captured as an answer
    for pattern, intent in re_patterns.items():
        if re.search(pattern, user_input, re.IGNORECASE):
            return intent

    if user_state["name"] and user_state["current_question"] < len(quiz_questions):
        # strip trailing qualifiers so "Tehran i guess" → "Tehran"
        cleaned = re.sub(r"\s*\b(i guess|right\??|i think so|you know)\b\s*$", "", user_input, flags=re.IGNORECASE)
        # strip apostrophes — they break [A-Za-z\s\-] and split "it's" into "it" + "s"
        cleaned = re.sub(r"'", "", cleaned)
        result = re.search(guess_pattern, cleaned, re.IGNORECASE)
        if result:
            user_state["last_guess"] = result.group("guess").strip()
            return "quiz_guess"

    return "default"

def check_answer(guess, question):
    correct = [question["answer"]] + question["aliases"]
    return any(c in guess.lower() for c in correct)

def chatbot_response(user_input):
    intent = detect_pattern(user_input)
    name = user_state["name"] or "friend"
    score = user_state["score"]
    q_idx = user_state["current_question"]
    total = len(quiz_questions)

    if intent == "name_response":
        response = random.choice(chatbot_responses["name_response"]).format(name=name, total=total)
        return response + "\n" + quiz_questions[0]["question"]

    elif intent == "quiz_guess":
        question = quiz_questions[q_idx]
        if check_answer(user_state["last_guess"], question):
            user_state["score"] += 1
            feedback = random.choice(chatbot_responses["correct"]).format(name=name, score=user_state["score"])
        else:
            feedback = random.choice(chatbot_responses["wrong"]).format(name=name, answer=question["answer"])
        user_state["answered"] += 1
        user_state["current_question"] += 1
        next_idx = user_state["current_question"]
        if next_idx < total:
            return feedback + "\n" + quiz_questions[next_idx]["question"]
        else:
            return feedback + "\n" + random.choice(chatbot_responses["quiz_complete"]).format(
                name=name, score=user_state["score"], total=total)

    elif intent == "hint_response":
        if user_state["name"] and q_idx < total:
            return random.choice(chatbot_responses["hint_response"]).format(hint=quiz_questions[q_idx]["hint"])
        return random.choice(chatbot_responses["default"])

    elif intent == "score":
        return random.choice(chatbot_responses["score"]).format(
            name=name, score=score, answered=user_state["answered"])

    elif intent in chatbot_responses:
        return random.choice(chatbot_responses[intent])

    else:
        return random.choice(chatbot_responses["default"])

# --- Main loop ---

def chat():
    # Reset state for a fresh game
    user_state.update({"name": None, "score": 0, "current_question": 0, "answered": 0, "last_guess": None})

    print("Welcome to the Iran Quiz Chatbot!")
    print("Type 'exit' or 'quit' to stop at any time.\n")
    print("Chatbot: Hi there! What is your name?")

    while True:
        user_input = input("You: ")

        if re.search(r"\b(exit|quit|bye|goodbye)\b", user_input, re.IGNORECASE):
            name  = user_state["name"] or "friend"
            score = user_state["score"]
            total = len(quiz_questions)
            print(f"Chatbot: Goodbye {name}! Final score: {score}/{total}. See you next time!")
            break

        bot_response = chatbot_response(user_input)
        print(f"Chatbot: {bot_response}")

chat()
```

```python
import json
import re
import random

NAMED_GROUP_TAGS = {"name", "quiz_guess"}

def load_files(filepath="intents.json"):
    """Load intents.json and build the data structures the chatbot needs.
    Returns rgx2int, int2res, the two named-group patterns, and quiz questions.
    """
    with open(filepath, "r") as f:
        data = json.load(f)

    rgx2int = {}
    int2res = {}
    intent_map = {}

    for intent in data["intents"]:
        tag = intent["tag"]
        patterns = intent["patterns"]
        intent_map[tag] = intent
        int2res[tag] = intent["responses"]
        if patterns and tag not in NAMED_GROUP_TAGS:
            # \b word boundaries prevent "hi" matching inside "hint", "history", etc.
            rgx2int[r"|".join(r"\b" + p + r"\b" for p in patterns)] = tag

    name_pat = intent_map["name"]["patterns"][0]
    guess_pat = intent_map["quiz_guess"]["patterns"][0]

    quiz_questions = data["quiz"]
    return rgx2int, int2res, name_pat, guess_pat, quiz_questions

# Quick check
rgx2int, int2res, name_pat, guess_pat, quiz_questions = load_files()
print(f"rgx2int: {len(rgx2int)} pattern groups → {list(rgx2int.values())}")
print(f"int2res: {list(int2res.keys())}")
print(f"quiz_questions: {len(quiz_questions)} questions loaded")
print(f"\nname_pat: {name_pat}")
print(f"guess_pat: {guess_pat[:60]}...")
```

```python
# generate_response: simplified version showing the core pattern-matching loop.
# The full chatbot below extends this to handle quiz logic and named-group patterns.

def generate_response(user_input, rgx2int, int2res):
    """Match user input against loaded patterns and return a random response."""
    for pattern, tag in rgx2int.items():
        if re.search(pattern, user_input, re.IGNORECASE):
            responses = int2res[tag]
            if responses:
                return random.choice(responses)
    return random.choice(int2res.get("default", ["I didn't understand that."]))

# Test the basic matcher
rgx2int_t, int2res_t, _, _, _ = load_files()
print(generate_response("hello there", rgx2int_t, int2res_t))
print(generate_response("what is my score", rgx2int_t, int2res_t))
print(generate_response("give me a hint", rgx2int_t, int2res_t))
print(generate_response("xyz unknown", rgx2int_t, int2res_t))
```

```python
import json
import re
import random

user_state = {"name": None, "score": 0, "current_question": 0, "answered": 0, "last_guess": None}

def detect_pattern(user_input, rgx2int, name_pat, guess_pat, quiz_questions):
    """Return the intent tag for user_input.
    Checks in order: name capture, side-intents, then quiz guess.
    """
    # 1. Named group: user's name
    result = re.search(name_pat, user_input)
    if result:
        user_state["name"] = result.group("name")
        return "name"

    # 2. Side-intent check first — so "hint", "score" etc. are never mistaken for a guess
    for pattern, tag in rgx2int.items():
        if re.search(pattern, user_input, re.IGNORECASE):
            return tag

    # 3. Named group: quiz guess (only while a question is pending)
    if user_state["name"] and user_state["current_question"] < len(quiz_questions):
        # strip trailing qualifiers so "Tehran i guess" → "Tehran"
        cleaned = re.sub(r"\s*\b(i guess|right\??|i think so|you know)\b\s*$", "", user_input, flags=re.IGNORECASE)
        # strip apostrophes — they break [A-Za-z\s\-] and split "it's" into "it" + "s"
        cleaned = re.sub(r"'", "", cleaned)
        result = re.search(guess_pat, cleaned, re.IGNORECASE)
        if result:
            user_state["last_guess"] = result.group("guess").strip()
            return "quiz_guess"

    return "default"

def check_answer(guess, question):
    """Return True if guess matches the correct answer or any alias."""
    correct = [question["answer"]] + question["aliases"]
    return any(c in guess.lower() for c in correct)

def chatbot_response(user_input, rgx2int, int2res, name_pat, guess_pat, quiz_questions):
    """Generate a response for user_input using data loaded from intents.json."""
    tag = detect_pattern(user_input, rgx2int, name_pat, guess_pat, quiz_questions)
    name = user_state["name"] or "friend"
    total = len(quiz_questions)
    q_idx = user_state["current_question"]

    if tag == "name":
        greeting = random.choice(int2res["name"]).format(name=name, total=total)
        return greeting + "\n" + quiz_questions[0]["question"]

    elif tag == "quiz_guess":
        question = quiz_questions[q_idx]
        if check_answer(user_state["last_guess"], question):
            user_state["score"] += 1
            feedback = random.choice(int2res["correct"]).format(
                name=name, score=user_state["score"], total=total)
        else:
            feedback = random.choice(int2res["wrong"]).format(
                name=name, answer=question["answer"])
        feedback += f"\nFact: {question['fact']}"
        user_state["answered"] += 1
        user_state["current_question"] += 1
        next_idx = user_state["current_question"]
        if next_idx < total:
            return feedback + "\n\n" + quiz_questions[next_idx]["question"]
        else:
            ending = random.choice(int2res["quiz_complete"]).format(
                name=name, score=user_state["score"], total=total)
            return feedback + "\n\n" + ending

    elif tag == "hint":
        if user_state["name"] and q_idx < total:
            return int2res["hint"][0].format(hint=quiz_questions[q_idx]["hint"])
        return random.choice(int2res["default"])

    elif tag == "score":
        return random.choice(int2res["score"]).format(
            name=name, score=user_state["score"], answered=user_state["answered"])

    elif tag in int2res and int2res[tag]:
        template = random.choice(int2res[tag])
        try:
            return template.format(name=name, score=user_state["score"], total=total)
        except KeyError:
            return template

    else:
        return random.choice(int2res["default"])

def chat():
    """Part 3 chatbot — all patterns and responses loaded from intents.json."""
    user_state.update({"name": None, "score": 0, "current_question": 0, "answered": 0, "last_guess": None})

    rgx2int, int2res, name_pat, guess_pat, quiz_questions = load_files()

    print("Welcome to the Iran Quiz Chatbot!")
    print("Type 'exit' or 'quit' to stop at any time.\n")
    print("Chatbot: Hi there! What is your name?")

    while True:
        user_input = input("You: ")

        if re.search(r"\b(exit|quit|bye|goodbye)\b", user_input, re.IGNORECASE):
            name = user_state["name"] or "friend"
            total = len(quiz_questions)
            print(f"Chatbot: Goodbye {name}! Final score: {user_state['score']}/{total}. See you next time!")
            break

        bot_response = chatbot_response(user_input, rgx2int, int2res, name_pat, guess_pat, quiz_questions)
        print(f"Chatbot: {bot_response}")

chat()
```

```python
# References:
# https://www.nltk.org/howto/stem.html
# https://www.nltk.org/api/nltk.stem.wordnet.html#nltk.stem.wordnet.WordNetLemmatizer
# https://www.nltk.org/api/nltk.stem.porter.html#nltk.stem.porter.PorterStemmer

import nltk
import re
import random
import json

nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

stop_words = set(stopwords.words('english'))
wnl = WordNetLemmatizer()

# 4.1 Preprocessing pipeline

def get_wordnet_pos(tag):
    """Map Penn Treebank POS tag to the label WordNetLemmatizer expects."""
    if tag.startswith('V'): return 'v'  # verb
    if tag.startswith('J'): return 'a'  # adjective
    if tag.startswith('R'): return 'r'  # adverb
    return 'n'                           # default: noun

def preprocess(text):
    """Tokenise, remove punctuation and stopwords, then lemmatize.
    Returns a list of normalised tokens.
    """
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w.isalpha()]
    tokens = [w for w in tokens if w.lower() not in stop_words]
    tagged = pos_tag(tokens)
    tokens = [wnl.lemmatize(w.lower(), get_wordnet_pos(t)) for w, t in tagged]
    return tokens

# Demo — shows why lemmatization improves matching over plain string comparison
print("=== Preprocessing demo ===")
demo = [
    ("Give me a hint please", "→ ['give', 'hint']  — still matches 'hint' intent"),
    ("scoring so far", "→ ['score']  — 'scoring' lemmatizes to 'score'"),
    ("I don't know!", "→ ['know']  — matches 'hint' intent via 'know'"),
    ("many thanks", "→ ['many', 'thank']  — 'thanks' lemmatizes to 'thank'"),
    ("could you help me out", "→ ['could', 'help']  — matches 'hint' intent via 'help'"),
]
for text, note in demo:
    tokens = preprocess(text)
    print(f"  IN:  '{text}'")
    print(f"  OUT: {tokens}  {note}")
    print()

# 4.2 Load files with preprocessed patterns

NAMED_GROUP_TAGS = {"name", "quiz_guess"}

def load_files(filepath="intents.json"):
    """Load intents.json — same as Part 3 but patterns are lemmatized
    so variant phrasings like 'scoring' still match 'score'.
    """
    with open(filepath, "r") as f:
        data = json.load(f)

    rgx2int = {}
    int2res = {}
    intent_map = {}

    for intent in data["intents"]:
        tag, patterns = intent["tag"], intent["patterns"]
        intent_map[tag] = intent
        int2res[tag] = intent["responses"]

        if not patterns or tag in NAMED_GROUP_TAGS:
            continue

        processed_pats = []
        for p in patterns:
            tokens = preprocess(p)
            if tokens:
                processed_pats.append(
                    r"\b" + r"\s+".join(re.escape(t) for t in tokens) + r"\b"
                )
        if processed_pats:
            rgx2int[r"|".join(processed_pats)] = tag

    name_pat = intent_map["name"]["patterns"][0]
    guess_pat = intent_map["quiz_guess"]["patterns"][0]
    quiz_questions = data["quiz"]
    return rgx2int, int2res, name_pat, guess_pat, quiz_questions

# 4.3 Chatbot functions

user_state = {"name": None, "score": 0, "current_question": 0, "answered": 0, "last_guess": None}

def detect_pattern(user_input, rgx2int, name_pat, guess_pat, quiz_questions):
    """Same three-stage logic as Part 3 but Stage 2 now matches
    preprocessed input against lemmatized patterns instead of raw strings.
    Stage 3 (quiz guess) is last resort so side-intents like 'hint' are never
    mistaken for an answer.
    """
    # Stage 1: name capture — raw input needed to preserve capitalisation [A-Z][a-z]+
    result = re.search(name_pat, user_input)
    if result:
        user_state["name"] = result.group("name")
        return "name"

    # Stage 2: side-intent check first (lemmatized) — "hint", "score" etc. matched before quiz guess
    tokens = preprocess(user_input)
    joined_input = " ".join(tokens)
    for pattern, tag in rgx2int.items():
        if re.search(pattern, joined_input, re.IGNORECASE):
            return tag

    # Stage 3: quiz guess — raw input, apostrophe-stripped (last resort)
    if user_state["name"] and user_state["current_question"] < len(quiz_questions):
        cleaned = re.sub(r"\s*\b(i guess|right\??|i think so|you know)\b\s*$", "", user_input, flags=re.IGNORECASE)
        cleaned = re.sub(r"'", "", cleaned)
        result = re.search(guess_pat, cleaned, re.IGNORECASE)
        if result:
            user_state["last_guess"] = result.group("guess").strip()
            return "quiz_guess"

    return "default"

def check_answer(guess, question):
    correct = [question["answer"]] + question["aliases"]
    return any(c in guess.lower() for c in correct)

def chatbot_response(user_input, rgx2int, int2res, name_pat, guess_pat, quiz_questions):
    tag = detect_pattern(user_input, rgx2int, name_pat, guess_pat, quiz_questions)
    name = user_state["name"] or "friend"
    total = len(quiz_questions)
    q_idx = user_state["current_question"]

    if tag == "name":
        greeting = random.choice(int2res["name"]).format(name=name, total=total)
        return greeting + "\n" + quiz_questions[0]["question"]

    elif tag == "quiz_guess":
        question = quiz_questions[q_idx]
        if check_answer(user_state["last_guess"], question):
            user_state["score"] += 1
            feedback = random.choice(int2res["correct"]).format(
                name=name, score=user_state["score"], total=total)
        else:
            feedback = random.choice(int2res["wrong"]).format(
                name=name, answer=question["answer"])
        feedback += f"\nFact: {question['fact']}"
        user_state["answered"] += 1
        user_state["current_question"] += 1
        next_idx = user_state["current_question"]
        if next_idx < total:
            return feedback + "\n\n" + quiz_questions[next_idx]["question"]
        else:
            ending = random.choice(int2res["quiz_complete"]).format(
                name=name, score=user_state["score"], total=total)
            return feedback + "\n\n" + ending

    elif tag == "hint":
        if user_state["name"] and q_idx < total:
            return int2res["hint"][0].format(hint=quiz_questions[q_idx]["hint"])
        return random.choice(int2res["default"])

    elif tag == "score":
        return random.choice(int2res["score"]).format(
            name=name, score=user_state["score"], answered=user_state["answered"])

    elif tag in int2res and int2res[tag]:
        template = random.choice(int2res[tag])
        try:
            return template.format(name=name, score=user_state["score"], total=total)
        except KeyError:
            return template

    else:
        return random.choice(int2res["default"])

# 4.4 Main loop

def chat():
    """Part 4 chatbot — NLTK lemmatization for robust, flexible intent detection."""
    user_state.update({"name": None, "score": 0, "current_question": 0, "answered": 0, "last_guess": None})

    rgx2int, int2res, name_pat, guess_pat, quiz_questions = load_files()

    print("Welcome to the Iran Quiz Chatbot!")
    print("Type 'exit' or 'quit' to stop at any time.\n")
    print("Chatbot: Hi there! What is your name?")

    while True:
        user_input = input("You: ")

        if re.search(r"\b(exit|quit|bye|goodbye)\b", user_input, re.IGNORECASE):
            name = user_state["name"] or "friend"
            total = len(quiz_questions)
            print(f"Chatbot: Goodbye {name}! Final score: {user_state['score']}/{total}. See you next time!")
            break

        bot_response = chatbot_response(user_input, rgx2int, int2res, name_pat, guess_pat, quiz_questions)
        print(f"Chatbot: {bot_response}")

chat()
```
